## VRE Curtailment
#### ((total available gen * hourly capacity factor profile) - dispatched (power.csv))/ available
#### Show as bar chart (6.4?)
#### use: curtail?

```python
# average capacity
capacity = pd.read_csv(results + 'capacity.csv')
cap = capacity.loc[capacity['Resource'].str.contains('wind|solar|pv'),:]
cap.index = cap['Resource']
cap = cap['EndCap'].T
cap.index = [elem.split('_')[0] for elem in list(cap.index)]

# capacity factor
cf = genvar.loc[:, genvar.columns.str.contains('wind|solar|pv')]
cf.columns = [elem.split('_')[0] for elem in list(cf.columns)]

# total generation
totalgen = (cap*cf).mul(time_weights['Weight'], 0)
totalgen = pd.DataFrame(totalgen.sum()) # sum across time
totalgen = totalgen.groupby(totalgen.index).sum() # sum across generators

# dispatched power
dispatch = power[2:].loc[:, power.columns.str.contains('wind|solar|pv')].reset_index(drop = True)
dispatch = dispatch.mul(time_weights['Weight'], 0)
dispatch.columns = [elem.split('_')[0] for elem in list(dispatch.columns)]
dispatch = pd.DataFrame(dispatch.sum()) # sum across time
dispatch = dispatch.groupby(dispatch.index).sum() # sum across generators

# curtailment calculation
curtailment = (1 - (dispatch)/(totalgen))*100

# plot
curtailment.set_index(curtailment.index).loc[regions].plot(kind="bar", legend=False)
plt.title('VRE Curtailment by Region' )
plt.xlabel('Regions')
plt.ylabel('VRE Curtailment (%)')
plt.show()
```

## LCOE

```python
lcoe = gendat.copy()
lcoe['Average_CF'] = list(genvar.iloc[:,1:].mean())
lcoe['Fuel_Cost_per_MMBTU'] = 0
for i in range(len(fuelcost)):
    lcoe.loc[lcoe.Fuel.str.contains(fuelcost.index[i]),'Fuel_Cost_per_MMBTU'] = fuelcost[i]
lcoe['LCOE_per_MWh_1.0'] = (((lcoe.Inv_Cost_per_MWyr + lcoe.Fixed_OM_Cost_per_MWyr) / (8760*lcoe.Average_CF)) + (lcoe.Fuel_Cost_per_MMBTU * lcoe.Heat_Rate_MMBTU_per_MWh) + lcoe.Var_OM_Cost_per_MWh)
lcoe['LCOE_per_MWh_0.5'] = (((lcoe.Inv_Cost_per_MWyr + lcoe.Fixed_OM_Cost_per_MWyr) / (8760*lcoe.Average_CF*0.5)) + (lcoe.Fuel_Cost_per_MMBTU * lcoe.Heat_Rate_MMBTU_per_MWh) + lcoe.Var_OM_Cost_per_MWh)
lcoe['LCOE_per_MWh_0.3'] = (((lcoe.Inv_Cost_per_MWyr + lcoe.Fixed_OM_Cost_per_MWyr) / (8760*lcoe.Average_CF*0.3)) + (lcoe.Fuel_Cost_per_MMBTU * lcoe.Heat_Rate_MMBTU_per_MWh) + lcoe.Var_OM_Cost_per_MWh)
lcoe_table = pd.DataFrame()
lcoe_table['1.0'] = lcoe.groupby(['Resource_Type'])['LCOE_per_MWh_1.0'].mean()
lcoe_table['0.5'] =lcoe.groupby(['Resource_Type'])['LCOE_per_MWh_0.5'].mean()
lcoe_table['0.3'] =lcoe.groupby(['Resource_Type'])['LCOE_per_MWh_0.3'].mean()
lcoe_table
```
